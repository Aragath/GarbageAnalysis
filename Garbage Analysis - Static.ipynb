### After knowing how the data look, we can concat CSV files of different months
12 datasets are now stored in the dfs dictionary, we can access different months' data by searching the key "month_{}"

```python
import glob

# Fetching file paths
file_paths = glob.glob('Datasets/*.csv')

dfs = {}

# Concat datasets from 12 months
for i, file_path in enumerate(file_paths):
    # Detect document encoding using chardet
    with open(file_path, 'rb') as f:
        result = chardet.detect(f.read())
    
    encoding = result['encoding']
    
    # Load dataset file
    _ = pd.read_csv(file_path, encoding=encoding)
    
    dfs[f'month_{i+1}'] = _

# Show concated dataset
dfs
```

### Analyzing: The trend of total amount of garbage throughout 2022

#### Total
From the line graph, we could observe that January has the most amout of garbage, followed by November and October.

We could also see that April and August has the least garbage throughout the year.

Let's take a deeper look into different genres of garbage to get to know more about what caused these highs and lows.

```python
pd.options.plotting.backend = "plotly"

total_amt = []
for i in range(1,13):
    total = pd.DataFrame(dfs[f'month_{i}']).iloc[0]['總計']
    total_amt.append(total)
    

df = pd.DataFrame(total_amt, columns=["總計"])
df.index += 1
fig = df.plot(labels=dict(index="月份", value="垃圾量"))
fig.show("png")
```

#### Paper
From the result, we could see that it has the same trend as the total amount of garbage, indicating the amount of paper garbage may have a great affect on total amount of garbage or many different kinds of garbage follows similar patterns.

```python
paper_list = []
for i in range(1,13):
    paper = pd.DataFrame(dfs[f'month_{i}']).iloc[1]['總計']
    paper_list.append(paper)
    
df = pd.DataFrame(paper_list)
df.index = list(range(1,13))
df.columns = ['paper']

# fig = df.plot()
# fig.show()
fig = df.plot(labels=dict(index="月份", value="垃圾量"))
fig.show("png")
```

#### Paper Container
From the result, we could also observe that paper containers has similar trend as paper garbage, hence we presume that there might be a similar trend in common.

```python
_ = []
for i in range(1,13):
    paper_container = pd.DataFrame(dfs[f'month_{i}']).iloc[2]['總計']
    _.append(paper_container)
    
df = pd.DataFrame(_)
df.index = list(range(1,13))
df.columns = ['paper_container']

fig = df.plot(labels=dict(index="月份", value="垃圾量"))
fig.show("png")
```

#### Let's visualize the trends of all kinds of garbage 
From the result, we could see that most genres follows similar trends, but there're still some special cases like Disk, Mobile Phone, Cooking Oil.
More, from the shared similar trends, we could indicate that the rise and drops are isn't genre specific and people just tend to throw things during that month or before/after particular event.

Disk: A significant rise in May and has no significant following drops

Mobile Phone (including charger) : Quite different from other genres, has irregular rises and drops

Cooking Oil: Similar trend as others but has a significant rise in May

```python
def generate_subplot_data(name, i):
    d_list = []
    for i in range(1,13):
        garbage = pd.DataFrame(dfs[f'month_{i}']).iloc[i]['總計']
        d_list.append(garbage)

    df = pd.DataFrame(d_list)
    df.index = list(range(1,13))
    df.columns = [name]
    return df
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(10, 20))

column_names= ['Total', 'Paper', 'Paper Container', 'Foil Pack', 'Foil Container', 'Iron Container', 'Other metals', 'Plastic Container',
               'Styrofoam for packaging', 'Other plastic', 'Tire', 'Glass Container', 'Other glass', 'Lighting', 'Dry Battery', 'Lead Battery',
               'Home Appliance', 'IT Objects', 'Disc', 'Mobile phone (including charger)', 'Pesticide/drug containers', 'Clothes', 'Cooking oil',
               'Others']

for i, ax in enumerate(axes.flat):
    data = generate_subplot_data(column_names[i], i)
    ax.plot(data)
    ax.set_xlabel('Month')
    ax.set_ylabel('Amount')
    ax.set_title(column_names[i])

plt.tight_layout()

plt.show()
```

### Analyzing: Comparing total amounts of garbage across different genre produced throughout 2022
After analyzing the total amount of garbage each month, I'm also curious about how different genre differ in cleaning methods

Doesn't look like there're much difference between genres across different months from the results, most garbages are recycled through public/private organizations.

But we could see that glass containers are the main garbage that environmental protection organization recycles.

```python
import plotly.graph_objects as go

for i in range(1,13):
    df = pd.DataFrame(dfs[f'month_{i}'])

    df = df.set_index('項目別')

    fig = go.Figure()
    for column in df.columns.difference(['總計']):
        fig.add_trace(go.Bar(name=column, x=df.index, y=df[column]))
    fig.update_layout(barmode='stack', title=f'{i}月垃圾分類統計')
    fig.update_xaxes(tickangle=-45)
    
    fig.show("png")
```

#### Visualization with interactive graphs
I've also built a interactive graph using dash, you could run this cell and switch through different graphs using a dropdown.

```python
import dash
from dash import dcc, html, Input, Output
import plotly_express as px

for i in range(1,13):
    df = pd.DataFrame(dfs[f'month_{i}'])

    df = df.set_index('項目別')

    globals()['fig'+str(i)] = go.Figure()
    for column in df.columns.difference(['總計']):
        globals()['fig'+str(i)].add_trace(go.Bar(name=column, x=df.index, y=df[column]))
    globals()['fig'+str(i)].update_layout(barmode='stack', title=f'{i}月垃圾分類統計')
    globals()['fig'+str(i)].update_xaxes(tickangle=-45)

    
# Build dash application
app = dash.Dash(__name__)

app.layout = html.Div(
 children = [
     dcc.Dropdown(
         id="dropdown",
         options=list(range(1,13)),
         value="1",
         clearable=False,
    ), 
     dcc.Graph(id='graph')
])

# build callback for dropdown
@app.callback(
    Output("graph", "figure"), 
    Input("dropdown", "value"))
def update_bar_chart(month):
    return globals()['fig'+str(month)]


# run application
if __name__ == "__main__":
    app.run_server(debug=False)
```

### Analyzing: Comparing the amount of resources across different recycle methods
At last, I also analyzed what percentage does each genre take between different recycle methods.

Doesn't look like there're much difference across different months from the results, but we could get to know the distribution of genres of garbage, and reconfirm that glass is the main garbage that environmental protection organization recycles.

```python
# Build graphs
for i in range(1, 13):
    df = pd.DataFrame(dfs[f'month_{i}']).transpose()
    df.columns = df.iloc[0]
    df = df[1:]

    fig = go.Figure()
    for column in df.columns.difference(['總計']):
        fig.add_trace(go.Bar(name=column, x=df.index, y=df[column]))

    fig.update_layout(barmode='stack', title=f'{i}月各項目垃圾清運量統計', xaxis_title='清運方式', yaxis_title='清運量')
    
    fig.show("png")
```
